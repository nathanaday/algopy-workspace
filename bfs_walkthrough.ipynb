# BFS (Breadth-First Search) Walkthrough

```python
import networkx as nx
from algopy.utilities import GraphVis
from algopy.algorithms import bfs as bfs_with_history

%matplotlib inline
```

## 1. Build the Graph

A connected, undirected graph with 10 nodes and enough edges to make
the traversal interesting.

```python
G = nx.connected_watts_strogatz_graph(10, 3, 0.4)
gv = GraphVis()
gv.display(G, title="Original Graph (10 nodes)")
```

## 2. Run BFS and Visualize Progress

```python
snapshots = list(bfs_with_history(G, start=0))
gv.show_progress(G, snapshots, title_prefix="BFS Step")
```

## 3. Final BFS Search Tree

```python
gv.show_tree(G, snapshots[-1][1], root=0, title="BFS Search Tree (rooted at node 0)")
```
