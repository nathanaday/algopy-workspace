# Min Heap Walkthrough

Visualize a min-heap as a binary tree using NetworkX.
We start empty, insert random values one at a time, then pop the min.

```python
import random
from algopy.utilities import draw_heap
from algopy.data_structures import MinHeap

%matplotlib inline
```

## 1. Empty Heap

```python
h = MinHeap()
draw_heap(h.to_networkx(), title="Empty Heap")
```

## 2. Insert Elements

Insert random integers one at a time and draw the heap after each insertion.

```python
random.seed(42)
values = random.sample(range(1, 100), 15)

for v in values:
    h.insert(f"node_{v}", v)
    draw_heap(h.to_networkx(), title=f"After insert('node_{v}', {v})  --  array: {[e['value'] for e in h.elements]}")
```

## 3. Pop the Min

Remove the root (minimum element) and show the heap after it re-heapifies.

```python
popped = h.pop()
print(f"Popped: {popped}")
draw_heap(h.to_networkx(), title=f"After pop() returned {popped['value']}  --  array: {[e['value'] for e in h.elements]}")
```
