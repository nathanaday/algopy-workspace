# DFS (Depth-First Search) Walkthrough

Build a connected, undirected graph with 10 nodes, run DFS step-by-step,
and visualize the algorithm's progress at key stages.

```python
import networkx as nx
from algopy.utilities import GraphVis
from algopy.algorithms import dfs as dfs_with_history

%matplotlib inline
```

## 1. Build the Graph


```python
G = nx.connected_watts_strogatz_graph(10, 3, 0.4)
gv = GraphVis()
gv.display(G, title="Original Graph (10 nodes)")
```

## 2. Run DFS and Visualize Progress

```python
snapshots = list(dfs_with_history(G, start=0))
gv.show_progress(G, snapshots, title_prefix="DFS Step")
```

## 3. Final DFS Search Tree

```python
gv.show_tree(G, snapshots[-1][1], root=0, title="DFS Search Tree (rooted at node 0)")
```
